#### Step 1: Import the necessary module.


### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Write your code here
summary_df = pd.DataFrame({
    'Data Type': df.dtypes,
    'Non-Null Count': df.count(),
    'Missing_values': df.isnull().sum()
})
print(summary_df)
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
print(df.describe())
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
unique_countries = sorted(df['Country'].dropna().unique())
print('Unique countries found:')
print(unique_countries)

country_counts = df['Country'].value_counts()
print('Top 15 most frequent countries:', country_counts.head(15))
print('Least frequent countries:', country_counts.tail(15))
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
country_mapping = {
    'US': 'United States',
    'USA': 'United States',
    'United States of America': 'United States',
    'UK': 'United Kingdom',
    'U.K.': 'United Kingdom',
    'Federal Republic of Germany': 'Germany'
}
df['Country'] = df['Country'].replace(country_mapping)
ed_mapping = {
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor Degree',
    'Master’s degree (M.A., M.S., M.Eng., etc.)': 'Master Degree',
    'Other doctoral degree (Ph.D., Ed.D., etc.)': 'Doctorate Degree',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'High School',
    'Some college/university study without earning a degree': 'Undergraduate (Incomplete)'
}
df['EdLevel'] = df['EdLevel'].replace(ed_mapping)

print("Standardized Education Levels:")
print(df['EdLevel'].unique())
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
employment_encoded = pd.get_dummies(df['Employment'], prefix='Employment', dtype='int')
df = pd.concat([df, employment_encoded], axis=1)
df = df.drop(columns=['Employment'])
print(df.filter(like='Employment').head())
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
missing_values_summary = df.isna().sum().sort_values(ascending=False)
print('Top 10 columns with the highest number of missing values: ')
print(missing_values_summary.head(10))
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
salary_mean = df['ConvertedCompYearly'].mean()
df['ConvertedCompYearly']=df['ConvertedCompYearly'].fillna(salary_mean)
print(f'Missing values remaining in ConvertedCompYearly: {df['ConvertedCompYearly'].isna().sum()}')
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
remote_work = df['RemoteWork'].mode()[0]
print(f'The most frequent value is: {remote_work}')
df['RemoteWork'] = df['RemoteWork'].fillna(remote_work)
print(f'Missing values remaining in RemoteWork: {df['RemoteWork'].isna().sum()}')
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
min_value = df['ConvertedCompYearly'].min()
max_value = df['ConvertedCompYearly'].max()

df['ConvertedCompYearly_scaled'] = (df['ConvertedCompYearly'] - min_value) / (max_value - min_value)
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_scaled']].head())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import numpy as np
df['ConvertedCompYearly_log'] = np.log1p(df['ConvertedCompYearly'])
print(f'Sesgo original: {df['ConvertedCompYearly'].skew():.2f}')
print(f'Sesgo despues del logaritmo: {df['ConvertedCompYearly_log'].skew():.2f}')
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_log']].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
df['YearsCodePro'] = df['YearsCodePro'].replace({'Less than 1 year': 0, 'More than 50 years': 51})
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

limites = [-1, 2, 5, 15, float('inf')]
categorias = ['Junior', 'Mid-Level', 'Senior', 'Expert']

df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], bins=limites, labels=categorias)

print(df[['YearsCodePro', 'ExperienceLevel']].head(10))
print("\nCantidad por nivel de experiencia:")
print(df['ExperienceLevel'].value_counts())
```
